```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.metrics import confusion_matrix

# Load dataset
dataset = pd.read_csv('/home/omkar/LP3/ML/3. Neural Network/Churn_Modelling.csv', index_col='RowNumber')

# Select columns
X = dataset.iloc[:,2:12]
Y = dataset.iloc[:,-1:]
```

```python
X = pd.get_dummies(X,columns=['Gender','Geography'])

X.head()


```

```python
# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Initialize ANN
classifier = Sequential()

# Add input layer and hidden layers
classifier.add(Dense(6, activation='relu', input_shape=(X_train.shape[1],)))
classifier.add(Dropout(rate=0.1))

# Add second hidden layer
classifier.add(Dense(6, activation='relu'))
classifier.add(Dropout(rate=0.1))

# Add output layer
classifier.add(Dense(1, activation='sigmoid'))

# Compile the model
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display the summary of the model
classifier.summary()

# Fit the Neural Network
history = classifier.fit(X_train, y_train, batch_size=32, epochs=20, validation_split=0.1, verbose=2)

# Predictions
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Accuracy of the NN
accuracy = ((cm[0][0] + cm[1][1]) * 100) / len(y_test)
print(f'{accuracy:.2f}% of data was classified correctly')
```
