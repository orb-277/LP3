```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train = train.drop(['PassengerId','Name','Ticket','Cabin','Embarked','Age'],axis =1)
test = test.drop(['PassengerId','Name','Ticket','Cabin','Embarked','Age'],axis =1)
```

```python
train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
train.iloc[:,2] = LabelEncoder().fit_transform(train.iloc[:,2])
train
```

```python
test.iloc[:,1] = LabelEncoder().fit_transform(test.iloc[:,1])
test
```

```python
train = train[['Sex','SibSp','Parch','Pclass','Fare','Survived']]
test = test[['Sex','SibSp','Parch','Pclass','Fare',]]
```

```python
x_tr = train.iloc[:,0:5]
y_tr = train.iloc[:,5]
x_tr
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_tr['Fare'] = sc.fit_transform(x_tr[['Fare']])
test['Fare'] = sc.fit_transform(test[['Fare']])

x_test = test


```

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
x_tr
```

```python
classifier = Sequential()


classifier.add(Dense(units=5, input_dim=5, kernel_initializer='uniform', activation='relu'))
classifier.add(Dense(units=3, kernel_initializer='uniform', activation='relu'))
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(x_tr, y_tr, batch_size = 16, epochs = 100)
```

```python
prediction = classifier.predict(test).tolist()
```

```python
y_test = pd.read_csv('gender_submission.csv')
```

```python
y_test2 = pd.read_csv('gender_submission.csv')
```

```python
y_test = y_test.iloc[:,1]
y_test
```

```python
predicted_labels = [0 if val[0] <= 0.5 else 1 for val in prediction] 
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, predicted_labels)
precision = precision_score(y_test, predicted_labels)
recall = recall_score(y_test, predicted_labels)
f1 = f1_score(y_test, predicted_labels)
confusion = confusion_matrix(y_test, predicted_labels)

# Print the metrics
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
print('Confusion Matrix:')
print(confusion)
```
